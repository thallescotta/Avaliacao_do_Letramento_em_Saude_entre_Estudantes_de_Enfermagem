```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Definindo o caminho para o arquivo Excel
file_path = '/mnt/data/(LGPD) Questionário de _Letramento em Saúde_ para Estudantes de Enfermagem (respostas).xlsx'

# Carregando o conjunto de dados do arquivo Excel
data = pd.read_excel(file_path)

# Selecionar apenas colunas com dados numéricos para cálculo da média
# Assumindo que as colunas com dados numéricos começam da 7ª coluna em diante
# Esta parte pode precisar ser ajustada dependendo da estrutura exata dos seus dados
colunas_numericas = data.select_dtypes(include=[np.number]).columns.tolist()
media_respostas = data[colunas_numericas].mean()

# As categorias usadas no gráfico radar devem corresponder às colunas selecionadas
categorias = colunas_numericas

# Adaptando as médias ao formato anteriormente utilizado
medias_novas = media_respostas.values.tolist()

# Repetindo o primeiro valor no final para fechar o círculo no gráfico radar
medias_novas += medias_novas[:1]

# Configurações do gráfico
num_vars = len(categorias)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Criando o gráfico radar com os novos dados
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, medias_novas, color='blue', alpha=0.25)
ax.plot(angles, medias_novas, color='blue', linewidth=2)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categorias, fontsize=12)

# Título
plt.title('Avaliação do Letramento em Saúde entre Estudantes de Enfermagem', size=15, color='blue', y=1.1)

plt.show()
```
